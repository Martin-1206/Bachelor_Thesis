```python
''' link: torso --> left_hip_roll ''' # Joint: hip_abduction_left
v01 = np.array([-0.001, 0.091, 0])
angle1 = -np.pi / 2
axis1 = np.array([0, 1, 0])
angle2 = -0.3752458
axis2 = np.array([0, 0, 1])
q01_1 = qmt.quatFromAngleAxis(angle1, axis1)
q01_2 = qmt.quatFromAngleAxis(angle2, axis2)
q01 = qmt.qmult(q01_1, q01_2)
print(q01)
#qmt.rotate(q01, v01)

print(qmt.quatFromEulerAngles(np.array([0, angle1, angle2]), axes='xyz')) #### ros macht xyz anscheinend

"""
Werte korrekt
Achtung: ROS macht xyzw, qmt ist wxyz !!!
hier stimmt was gehörig nicht in ros ist eine rotation mehr aber werte trotzdem richtig
"""
```

```python
''' left_hip_roll --> left_hip_yaw  ''' # joint: hip_rotation_left
v12 = np.array([-0.0505, 0, 0.044])     # Vector / Translation
angle = -np.pi / 2                       # Winkel um den gedreht wird
axis = np.array([0, 1, 0])              # Achse um die gedreht wird
q12 = qmt.quatFromAngleAxis(angle, axis)    # erzeugt Rotationsquaternion
#qmt.rotate(q12, v12)   # rotieren
print('w x y z')
print(q12)
""" 
Werte korrekt
Achtung: ROS macht xyzw, qmt ist wxyz
"""

qmt.qmult(q01, q12)
```

```python
def get_quat_23(q):
    var1 = 0.270598
    var2 = 2
    var3 = np.cos(q/var2)
    var4 = -0.653281
    var5 = np.sin(q/var2)
    var6 = 0.653281
    var7 = var1

    return np.array(
        [
            
            ((var7*var3) + (var6*var5)),
            ((var1*var3) - (var4*var5)),
            ((var1*var5) + (var4*var3)),
            ((var3*var6) - (var7*var5)),
            
        ]
    )
```
