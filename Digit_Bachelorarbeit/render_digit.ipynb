```python
import os
import mujoco_viewer
import mujoco._render
import numpy as np
import qmt
```

```python
model = mujoco.MjModel.from_xml_path('/home/martin/Bachelorarbeit/DIGIT/digit-v3_stl/digit-v3.xml') # create the Model from Path
data = mujoco.MjData(model) # creates the data
```

```python
viewer = mujoco_viewer.MujocoViewer(model, data)

# simulate and render
for _ in range(10000):
    if viewer.is_alive:
        
        
        # control motor
        data.ctrl[0] = 500 * (np.deg2rad(10.0) - data.sensordata[model.sensor_adr[mujoco.mj_name2id(model, 19, "left-hip-roll" )]])
        data.ctrl[6] = 500 * (np.deg2rad(0.0) - data.sensordata[model.sensor_adr[mujoco.mj_name2id(model, 19, "right-hip-roll" )]]) # in der simulation 1/80 torque drauf geben, gear ratio

        #print(data.sensordata[model.sensor_adr[mujoco.mj_name2id(model, 19, "left-hip-roll")]])
        #print('vel: ', data.sensordata[model.sensor_adr[mujoco.mj_name2id(model, 19, "left-hip-roll-vel")]])

        mujoco.mj_step(model, data)
        viewer.render()
    else:
        break

# close
viewer.close()
```
